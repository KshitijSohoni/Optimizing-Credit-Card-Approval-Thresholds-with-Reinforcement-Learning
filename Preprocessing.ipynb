```python
missing_values = df.isnull().sum()
missing_proportion = missing_values / len(df) * 100
missing_info = pd.DataFrame({'Missing Values': missing_values, 'Proportion (%)': missing_proportion})
print(missing_info[missing_info['Missing Values'] > 0])
```

```python
df=df.drop(['PROFESSIONAL_CITY'], axis=1)
```

```python
missing_values = df.isnull().sum()
missing_proportion = missing_values / len(df) * 100
missing_info = pd.DataFrame({'Missing Values': missing_values, 'Proportion (%)': missing_proportion})
print(missing_info[missing_info['Missing Values'] > 0])
```

```python
df = df.dropna(subset=['RESIDENCE_TYPE'])
df = df.dropna(subset=['MONTHS_IN_RESIDENCE'])
df = df.dropna(subset=['OCCUPATION_TYPE'])

df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
cat_var = df.select_dtypes(exclude=['number']).columns
le = LabelEncoder()
for i in cat_var:
    df[i] = le.fit_transform(df[i].astype(str))

df.head()
```

```python
df.to_csv('final_data.csv', index=False)
```
